```python
from gtts import gTTS
import os

# Input text
text = "Hello, this is a text-to-speech test."

# Language code (e.g., 'en' for English, 'es' for Spanish)
language = 'en'

# Slow speech speed
slow_speed = False

# Top-Level Domain to select accent (e.g., 'com.au' for Australian English)
tld = 'co.in'  # British English accent

# Create gTTS object with specified parameters
speech = gTTS(text=text, lang=language, slow=slow_speed, tld=tld)

# Save the converted audio to an mp3 file
speech.save("output.mp3")

# Play the converted file
os.system("start output.mp3")  # On Windows, use 'start'; on macOS, use 'afplay'; on Linux, use 'mpg321' or 'aplay'
```
